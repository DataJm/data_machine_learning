# Lista

* Se construyen usando los corchetes cuadrados []
* Son ordenadas, mutables y los elementos pueden ser cualquier cosa

Puedes combinar diferentes tipos de datos: strings, enteros, otras listas, etc.



```python
lista = [1,2,3,4,5]
```

```python
type(lista)
```

# Array

Tambien son colecciones de items

* Son ordenados, mutables, se construyen con algún modulo externo ( usaremos Numpy)
* En general no permite combinar diferentes tipos de datos (pero eso depende del tipo de array)
* Los array son fundamentales en el computo científico, muchos modelos parten de utilizar arrays. 

```python
import numpy as np
```

```python
array_1 = np.array([1,2,3,4,5])
```

```python
array_1
```

```python
type(array_1)
```

```python
array_1 / 2
```

# Series

Pandas está construido con Numpy, Python y C. En pandas podemos trabajar con diferentes fuentes de datos: JSON, Excel, SQL, etc. Esta pensado para facilitar operaciones de transformación de datos. 

```python
import pandas as pd
```

```python
serie = pd.Series([1,2,3,4,5])
```

```python
type(serie)
```

# En conclusión:

* Mutabilidad: Las 3 estructuras de datos son mutables.
* Homogeneidad: Las listas y las Series son heterogeneas, es decir, puedes tener diferentes tipos de datos. En los Array tenemos el mismo tipo de dato. 
* Accesibilidad: La lista y el array es por posición; en la serie es por index. 
* Las listas son parte de Python 
* El Array está pensando en ser eficiente en el computo estadístico.
* La serie está pensada en el tratamiendo y manipulacíon de datos.

# Repaso de objetos en python (clases)

```python
class Persona():
    def __init__(self, name, edad):
        self.name = name
        self.edad = edad
    def saluda(self):
        print(f"Hola amigos, yo soy {self.name}, tengo de edad: {self.edad}")
```

```python
persona1 = Persona("jose",35)
persona2 = Persona("Marco",30)
```

```python
persona1.saluda()
```

```python
persona2.saluda()
```
